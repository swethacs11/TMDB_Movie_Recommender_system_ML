```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
#pick only first three actors. pick first three 'name' from cast
```

```python
def convert3(obj):
    l=[]
    c=0
    for i in ast.literal_eval(obj):
        if c!=3:
            l.append(i['name'])
            c+=1
        else:
            break
    return l
```

```python
convert3(movies['cast'][1])
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
#pick director name from crew.
#pick the name where job isdirector
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i['job']=='Director'):
            L.append(i['name'])
            break
    return L
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#example 1

print(ps.stem('loved'))
print(ps.stem('love'))
print(ps.stem('lovely'))
print(ps.stem('loving'))
```

```python
#example 2

print(ps.stem('danced'))
print(ps.stem('dance'))
print(ps.stem('dancing'))
```

```python
def stem(text):
    y=[]
    string=""
    for i in text.split(): #split is used to convert string to list
        y.append(ps.stem(i))
        string = " ".join(y) #convert list back to string
    return string
```

```python
new_df['tags'][0]
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=500,stop_words='english') 
#max_features is how many most common words are considered. 
#stop_words describes english stop words must be removed.
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0] #this gives the first movie.
```

```python
cv.get_feature_names() #this shows which 5000 common words are selected.
```

```python
#we have 4806 movies so 4806 vectors.
# find distance between each vector with every other vector to find the similarity.
#we use the angle between vectors to find the distance. 
#if angle is 0 then movie is same. if 180 then very dissimilar movies etc.
#cosine distance is used.
#distance is inversely proportional to similarity.
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#similarity ranges from 0 to 1. 
#if 1 highly similar.
#if 0 highly dissimilar
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
similarity[0] #this gives the similarity value of 1st movie with every other movie.
```

```python
#how to find index of a movie
new_df[new_df['title']=="Batman Begins"].index[0]
```

```python
#def recommend(movie):
movie_index=new_df[new_df['title']=="Avatar"].index[0]
distances=similarity[movie_index]
    
    
```

```python
distances #this shows the similarity of batman begins with every other movie.
#if we sort and find the last five movies, we can get the similar movies but the index is lost.
```

```python
#to hold the movie index we will use enumerate.
#enumerate gives the enumerate object that has indexing from 0.
list(enumerate(distances))
```

```python
#lets sort it in descending order.
sorted(list(enumerate(distances)),reverse=True)
# this is sorting based on index. but we need to sort based on similarity.
```

```python
#sorting based on the similarity and pick only 5 most similar movie. leaving the similarity of movie with itself.
sorted(list(enumerate(distances)),reverse=True,key=lambda x : x[1])[1:6]
```

```python
#how to use iloc
print(new_df.iloc[322].title)
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movies=sorted(list(enumerate(distances)),reverse=True,key=lambda x : x[1])[1:6]
    for i in movies:
        print(new_df.iloc[i[0]].title) #i[0] has movie index.
        
    
```

```python
recommend("Avatar")
```
